```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("electricity_bill_dataset.csv")
print(df.head())
```

```python
X = df[['Units', 'Appliances', 'Hours_per_day']]
y = df['Bill']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1
)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
new_data = [[250, 5, 6]]
predicted_bill = model.predict(new_data)

print("Predicted Electricity Bill: ₹", predicted_bill[0])
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Bill")
plt.ylabel("Predicted Bill")
plt.title("Actual vs Predicted Electricity Bill")
plt.show()
```
